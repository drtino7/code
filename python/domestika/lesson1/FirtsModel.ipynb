```python
import pandas as pd
Data = pd.read_csv("data.csv")
```

```python
Data.info()
Data.head()
```

```python
import seaborn as sb
sb.scatterplot(x="celcius",y="fahrenheit",data=Data,hue="celcius",palette="coolwarm")
```

```python
X=Data["celcius"]
Y=Data["fahrenheit"]

Data_x = X.values.reshape(-1,1)
Data_y = Y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

Model=LinearRegression()
```

```python
Model.fit(Data_x,Data_y)
```

```python

#i give X it return Y
#i give caracteristics it returns tags

Model.predict([[77]])

Model.score(Data_x,Data_y)
```
