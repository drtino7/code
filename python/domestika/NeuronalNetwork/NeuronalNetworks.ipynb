```python
import pandas as pd
data = pd.read_csv("data.csv")
data.head()
```

```python
x = data["celcius"].values
y = data["fahrenheit"].values
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), #input
    tf.keras.layers.Dense(units=1)  #output
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss="mean_squared_error"
)
```

```python
train = model.fit(x,y,epochs=1000)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("# epochs")
plt.ylabel("# loss")
plt.plot(train.history["loss"])
```

```python
celcius = 123
predict = model.predict([celcius])
print(predict)
```
